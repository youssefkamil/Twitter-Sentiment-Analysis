```python
import re
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import itemfreq
import collections
```

```python
cols=['sentiment','id','date','query','user','tweet']
df = pd.read_csv("C:\\Users\\THE ONE\\Desktop\\MARKOV\\trainingandtestdata\\train.csv",encoding='latin1',names=cols)
df = df.iloc[:,[0,5]]
```

```python
df.head()
```

```python
np.random.seed(0)
index=np.random.randint(low=0, high=1599999,size=160000)
df=df.iloc[index,:]
df= df.reset_index(drop=True)
df['sentiment']=df['sentiment'].replace({4:1})
```

```python
df
```

```python
pat1='@[^ ]+'
pat2='[0-9]'
pat3='#[^ ]+'
pat4='www.[^ ]+'
pat5='http[^ ]+'
pat6='i'

combined_pat = '|'.join((pat1,pat2,pat3,pat4,pat5))
```

```python
clean = []
for t in df['tweet']:
    lower=t.lower()
    re_out= re.sub(combined_pat,' ',lower)
    neg_out=re.sub("n't",'not',re_out)
    clean.append(neg_out)
```

```python
clean
```

```python
clean_df = pd.DataFrame(clean,columns=['text'])
```

```python
clean_df ['sentiment'] = df['sentiment']
clean_df
```

```python
x=clean_df["text"]
y=clean_df["sentiment"]
```

```python
y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
x_train
```

```python
cv=CountVectorizer(stop_words='english',binary=False,ngram_range=(1,3),)
```

```python
cv.fit(x_train)
```

```python
cv.vocabulary_
```

```python
len(cv.get_feature_names())
```

```python
x_train_cv=cv.transform(x_train)
x_test_cv=cv.transform(x_test)
```

```python
x_train_cv.toarray
```

```python
lr=LogisticRegression()
lr.fit(x_train_cv,y_train)
```

```python
y_pred_train=lr.predict(x_train_cv)
print("Accuracy : ", accuracy_score(y_pred_train,y_train) )

y_pred=lr.predict(x_test_cv)
print("Accuracy : ", accuracy_score(y_pred,y_test) )
```

```python
from keras import Sequential
from keras.layers import Dense,Dropout,LSTM,Embedding
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
model_1=Sequential()

model_1.add(Dense(1,activation='relu', input_dim=x_train_cv.shape[1]) )
model_1.add(Dropout(0.4))
model_1.add(Dense(32,activation='relu', input_dim=x_train_cv.shape[1]) )
model_1.add(Dropout(0.4))
model_1.add(Dense(1,activation='sigmoid') )

model_1.compile('adam',loss='binary_crossentropy',metrics=['accuracy'])

model_1.summary()
```

```python
model_1.fit(x_train_cv,y_train,batch_size=16,epochs=5,validation_data=(x_test_cv,y_test))
```

```python
from keras.models import model_from_json,load_model

model_json= model_1.to_json()
with open ("model_1.json","w") as json_file:
    json_file.write(model_json)
model_1.save_weights("model_1 weights.h5")
model_1.save("model_1.h5")


```

```python
stop_words=stopwords.words('english')
x_train_rnn=[]
for text in x_train:
    toknized_words=word_tokenize(text)
    non_stop_words=[i for i in toknized_words if i not in stop_words]
    tweet=" ".join(non_stop_words)
    x_train_rnn.append(tweet)
```

```python
tk=Tokenizer()
tk.fit_on_texts(x_train_rnn)
```

```python
x_train_tok=tk.texts_to_sequences(x_train_rnn)
x_test_tok=tk.texts_to_sequences(x_test)
```
